```python
# Dependencies and Setup
import numpy as np
import pandas as pd

# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(csv_path)
data.head()
```

## Player Count

* Display the total number of players


```python
# Unique number of players
players = data["SN"].unique()
total = len(players)
total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Use mean for average, unique for Item ID to avoid repeated ids and its length for ID count, count Purchase ID for total transactions and sum Price for total revenue 
average = data["Price"].mean()
item_id = data["Item ID"].unique()
item = len(item_id)
purchase = data["Purchase ID"].count()
revenue = data["Price"].sum()
#print(average,item,purchase,revenue)
analysis = pd.DataFrame({"Unique Items":[item],"Average Price":[average], "Number of Purchases":[purchase],"Total Revenue":[revenue]})
#Formatting
analysis["Average Price"] = analysis["Average Price"].map("${:.2f}".format) 
analysis["Total Revenue"] = analysis["Total Revenue"].map("${:,.2f}".format)
analysis
```

##Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group by gender to get totals for male and female. Note: have to use unique to avoid repeated names as repeated gender counts
group = pd.DataFrame(data['Gender'].groupby(data['SN']).unique())
group
male = np.sum(group['Gender'] == "Male")
female = np.sum(group['Gender'] == "Female")
#Total players minus male and female must be equal to 'other'/'nondisclosed' player's gender
other = len(players) - male - female
total = male + female + other
#print(male,female,other,total) 
#Divide each (male, female and other) by total to get its percentage
p_male = male/total*100
p_female = female/total*100
p_other = other/total*100
#Construct DataFrame with Gender, Total Count per gender, and Percentage by Gender
demographics = pd.DataFrame({"Gender":["Male","Female","Other"], "Total Count":[male,female,other],"Percentage of Players":[p_male,p_female,p_other]})
#Formatting
demographics["Percentage of Players"] = demographics["Percentage of Players"].map("{:.2f}%".format) 
demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group data by Gender, sum prices to get total value
group = data.groupby("Gender")
total = group["Price"].sum()
average = group["Price"].mean()
count = group["Purchase ID"].count()
#Average per person is total per gender divided by number of male, female and other respectively
ppmale = data[data["Gender"] == "Male"].sum()["Price"] / male
ppfemale = data[data["Gender"] == "Female"].sum()["Price"] / female
ppother = data[data["Gender"] == "Other / Non-Disclosed"].sum()["Price"] / other
#print(ppmale, ppfemale, ppother) 
analysis2 = pd.DataFrame({
    "Total Purchases":count,
    "Average Price": average,
    "Total Purchase Value": total, 
    "Avg. Total Purchase per Person" : [ppmale, ppfemale, ppother]})
#Formatting
analysis2["Total Purchase Value"] = analysis2["Total Purchase Value"].map("${:,.2f}".format) 
analysis2["Average Price"] = analysis2["Average Price"].map("${:.2f}".format)
analysis2["Avg. Total Purchase per Person"] = analysis2["Avg. Total Purchase per Person"].map("${:.2f}".format)
analysis2
```

## Purchasing Analysis (Age)

```python
#print(data["Age"].max())
#print(data["Age"].min())
#Drop Duplicates to avoid repeated customer names 
age_unique = data.drop_duplicates("SN")
#Max value in bins based on 46 as max value was identified previously to be 45 
bins = [0,9.99,14,19,24,29,34,39,46]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
age_unique["Age Table"] = pd.cut(age_unique["Age"], bins, labels=group_names, include_lowest=True)
age_count = age_unique["Age Table"].value_counts()
age_percentage = age_count/len(age_unique)*100
purchase = pd.DataFrame({"Count by Age":age_count, "Percentage of Players":age_percentage})
#Formatting
purchase["Percentage of Players"] = purchase["Percentage of Players"].map("{:.2f}%".format) 
#sort by index (table) to have ascenidng order
purchase.sort_index()
```

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Same Bins and names as in previous excercise
bins = [0,9.99,14,19,24,29,34,39,46]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
data["Age Table"] = pd.cut(data["Age"], bins, labels=group_names)
analysis2 = data.groupby("Age Table")
purchase_count = data["Age Table"].value_counts()
purchase_total = analysis2["Price"].sum()
purchase_avg = analysis2["Price"].mean()
#Use Age Count from previous excercise to get unique number of persons per bin and calculate Avg. Purchase per Person
purchase_pp = purchase_total/age_count
purchase_pp
purchase2 = pd.DataFrame({
    "Purchase Count":purchase_count, 
    "Avg Purchase Price":purchase_avg, 
    "Total Purchase Value":purchase_total,
    "Avg Total Purchase per Person":purchase_pp})
#Formatting
purchase2["Total Purchase Value"] = purchase2["Total Purchase Value"].map("${:,.2f}".format) 
purchase2["Avg Purchase Price"] = purchase2["Avg Purchase Price"].map("${:.2f}".format)
purchase2["Avg Total Purchase per Person"] = purchase2["Avg Total Purchase per Person"].map("${:.2f}".format)
analysis2
#sort by index (table) to have ascenidng order
purchase2.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by name then count total purchases, average price and total value of purcases. 
name_group = data.groupby("SN")
name_pc = name_group["Purchase ID"].count()
name_avg = name_group["Price"].mean()
name_total = name_group["Price"].sum()
name_analysis = pd.DataFrame({
    "Purchase Count":name_pc, 
    "Avg Purchase Price":name_avg, 
    "Total Purchase Value":name_total})

#Sort by total value 
name_analysis.sort_values(["Total Purchase Value"], ascending=False, inplace = True)

#Formatting
name_analysis["Total Purchase Value"] = name_analysis["Total Purchase Value"].map("${:,.2f}".format) 
name_analysis["Avg Purchase Price"] = name_analysis["Avg Purchase Price"].map("${:.2f}".format)

#Display top 5 spenders
name_analysis.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Item ID and Item Name, both as indexes. Total Value is  the sum of prices, price is its mean, and count to get prurchase count. 
itemdf = data.groupby(["Item ID", "Item Name"])
tp = itemdf.sum()["Price"]
avg = itemdf.mean()["Price"]
count = itemdf.count()["Price"]
#Construct DataFrame then sort by Purchase Count and display top 5 items
item_summary = pd.DataFrame({
    "Purchase Count" : count,
    "Item Price": avg,
    "Total Purchase Value" : tp})
#item_summary

#Formatting
item_summary["Total Purchase Value"] = item_summary["Total Purchase Value"].map("${:.2f}".format) 
item_summary["Item Price"] = item_summary["Item Price"].map("${:.2f}".format)

#Sort by Purchase Count 
item_summary.sort_values(["Purchase Count"], ascending=False, inplace=True)

#Display top 5 items
item_summary.head(5)
```
